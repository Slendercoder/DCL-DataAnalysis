#### Heurística para escoger una casilla con base en Win_Stay

```python
def logistic(x, steepness, threshold):
    return 1 / (1 + np.exp(-steepness*(x - threshold)))

def win_stay_tile(score, t_n, gamma, delta):
    return round(0.5 + (2*t_n - 1) * 0.5 * logistic(score, gamma, delta), 2)

gamma = 2
delta = 27

fig, ax = plt.subplots(figsize=(6,6))
score = [32*round(0.01*i, 2) for i in range(0, 101)]

t_n = 0
y = [win_stay_tile(i, t_n, gamma, delta) for i in score]
sns.lineplot(x=score, y=y, ax=ax, label="Not previously selected")

t_n = 1
y = [win_stay_tile(i, t_n, gamma, delta) for i in score]
sns.lineplot(x=score, y=y, ax=ax, label="Previously selected")

ax.set_ylim([0,1])
ax.set_ylabel("Probability of selecting tile")
ax.set_xlabel("Score")
```

```python
def tile_closeness(k, r, t, alpha, beta, focales, estrategias, DEB=False):
    '''
    Retorna la probabilidad de que la casilla t sea seleccionada
    con base en la distancia entre la region r y la región focal k
    Input:
    - k, region focal
    - r, region seleccionada por el jugador en la ronda anterior como vector 64-bits
    - t, casilla en la rejilla
    - focales, lista con las regiones focales como vectores 64-bits
    - estrategias, lista con las regiones focales como lista de casillas
    '''
    f = focales[k]
    t_n = 1 if t in estrategias[k + 1] else 0
    distancia = np.abs(np.subtract(f, r)).sum()
    if DEB:
#        imprime_region(f)
#        imprime_region(r)
        print(f"Considerando region focal {nameRegion(k + 1)}.")
        if t_n == 0:
            print(f"Casilla {t} no está en {nameRegion(k + 1)}.")
        else:
            print(f"Casilla {t} sí está en {nameRegion(k + 1)}.")
        print(f"La distancia de la región a la focal es {distancia}")
    return round(0.5 * (t_n + logistic(distancia, -(2*t_n-1)*alpha, beta)), 2)
```

```python
k = 4
r_inicial = code2Vector(shaky_hand(estrategias[k+1], 6), 8)
r = r_inicial
f = focales[k]
#imprime_region(r)
alpha = 0.3
beta = 20

dists = []
for m in range(1):
    for n in range(15):
        estrategia = []
        for t in range(64):
            p = tile_closeness(k, r, t, alpha, beta, focales, estrategias)
    #        print(f"La probabilidad de seleccionar a {t} es {p}")
            if uniform(0,1) < p:
    #            print(f"Casilla {t} seleccionada!")
                estrategia.append(t)
        r = code2Vector(estrategia, 8)
    #    imprime_region(r)
        d = np.abs(np.subtract(f, r)).sum()
        dists.append(d)
#    print(f"Distancia inicial a {nameRegion(k)}: {np.abs(np.subtract(f, r_inicial)).sum()} Final: {d}")
print(dists)
print(f"Distancia inicial a {nameRegion(k)}: {np.abs(np.subtract(f, r_inicial)).sum()} Media final: {np.mean(dists)}")
    
```

### Uniendo las dos heurísticas

```python
def datos_conjunta(t, t_n):
    scores = []
    distancias = []
    probabilidad = []
    for s in range(20, 33):
        for d in range(20):
            scores.append(s)
            distancias.append(d)
            p_win_stay = win_stay_tile(s, t, gamma, delta)
            p_attr = round(0.5 * (t_n + logistic(d, -(2*t_n-1)*alpha, beta)), 2)
#            p = p_attr * p_win_stay
            winner = np.argmax([abs(p_win_stay - 0.5), abs(p_attr - 0.5)])
            if winner == 0:
                p = p_win_stay
            else:
                p = p_attr
            probabilidad.append(p)        
    df = pd.DataFrame({'X':scores, 'Y':distancias, 'Z':probabilidad})
    return df
```

```python
t = 1
t_n = 1
df = datos_conjunta(t, t_n)
df['p_ws'] = df.apply(lambda x: win_stay_tile(x['X'], t, gamma, delta), axis=1)
df['p_attr'] = df.apply(lambda x: round(0.5 * (t_n + logistic(d, -(2*t_n-1)*alpha, beta)), 2), axis=1)
df['winner'] = df.apply(lambda x: np.argmax([abs(x['p_ws'] - 0.5), abs(x['p_attr'] - 0.5)]), axis=1)
df.head(10)
```

#### Heurística para escoger una casilla con base en repulsión a complemento

```python
fig, ax = plt.subplots(figsize=(6,6))

alpha = 0.05
beta = 20
step = 100
distancias = [64/step * i for i in range(step)]

t_n = 0
y = [round(0.5 * (t_n + logistic(distancia, -(2*t_n-1)*alpha, beta)), 2) for distancia in distancias] 
sns.lineplot(x=distancias, y=y, ax=ax, label="Not in complement")

t_n = 1
y = [round(0.5 * (t_n + logistic(distancia, -(2*t_n-1)*alpha, beta)), 2) for distancia in distancias] 
sns.lineplot(x=distancias, y=y, ax=ax, label="In complement")

ax.set_ylim([-0.1,1.1])
ax.set_ylabel("Probability of selecting tile")
ax.set_xlabel("Distance between overlap and focal region")
```

```python
def tile_overlap(k, o, t, epsilon, zeta, focales, estrategias, DEB=False):
    '''
    Retorna la probabilidad de que la casilla t sea seleccionada
    con base en la distancia entre el overlap o y el complemento
    de la región focal k
    Input:
    - k, region focal
    - o, overlap entre los jugadores como vector 64-bits
    - t, casilla en la rejilla
    - focales, lista con las regiones focales como vectores 64-bits
    - estrategias, lista con las regiones focales como lista de casillas
    '''
    f = [1 - x for x in focales[k]]
    t_n = 0 if t in estrategias[k + 1] else 1
    distancia = np.abs(np.subtract(f, o)).sum()
    if DEB:
        print(f"Considerando region focal {nameRegion(k)}.")
        if t_n == 0:
            print(f"Casilla {t} no está en {nameRegion(k)}.")
        else:
            print(f"Casilla {t} sí está en {nameRegion(k)}.")
        print(f"La distancia de la región a la focal es {distancia}")
    return round(0.5 * (t_n + logistic(distancia, -(2*t_n-1)*epsilon, zeta)), 2)
```

### Estimando estrategia

```python
def region2strategy(region):
    strategy = []
    for i in range(64):
        if region[i] == 1:
            strategy.append(i)
    return strategy

def use_predictors(tile, region, score, overlap, pars, focals, strategies, DEB=False):
    
    # Obtener parametros para sigmoides
    alpha = pars[17]
    beta = pars[18]
    gamma = pars[19]
    delta = pars[20]
    epsilon = pars[21]
    zeta = pars[22]
    t_n = 1 if tile in region2strategy(region) else 0
    
    # Crear vector de probabilidades
    probs = [0.5] * 18
    for i in range(8):
        if pars[i]:
 #           print(f":::::::SE USA PREDICTOR DE {nameRegion(i + 1)} ({pars[i]})")
            p = tile_closeness(i, region, tile, alpha, beta, focals, strategies, DEB)
#            print(i, tile, alpha, beta, p)
            probs[i] = p
    for i in range(8):
        if pars[i + 8]:
            p = tile_overlap(i, overlap, tile, epsilon, zeta, focals, strategies)
            probs[i + 8] = p
    if pars[16]:
        probs[16] = win_stay_tile(score, t_n, gamma, delta)
    
    if DEB:
        print("Probabilidades:", probs)
    
    # Determinar probabilidad extrema
    extremos = [abs(x) for x in np.subtract(probs, [0.5] * 18)]
    indice = np.argmax(extremos)
    p = probs[indice]
    if DEB:
        print("Extremos:", extremos)
        print(f"Indice máximo {indice} con probabilidad {p}")
    
    return p

def estimate_strategy(region, score, overlap, parameters, focals, strategies, DEB=False):
    
    pars = [0]*23
    pars[0] = parameters[0]
    pars[1] = parameters[1]
    pars[2] = parameters[2]
    pars[3] = parameters[2]
    pars[4] = parameters[2]
    pars[5] = parameters[2]
    pars[6] = parameters[3]
    pars[7] = parameters[3]
    pars[8] = parameters[4]
    pars[9] = False # Overlap is never repelled away from NOTHING
    pars[10] = parameters[5]
    pars[11] = parameters[5]
    pars[12] = parameters[5]
    pars[13] = parameters[5]
    pars[14] = parameters[6]
    pars[15] = parameters[6]
    pars[16] = parameters[7]
    pars[17] = 1
    pars[18] = 15
    pars[19] = 2
    pars[20] = 27
    pars[21] = 0.05
    pars[22] = 20
    
    new_strategy = []
    for tile in range(64):
        p = use_predictors(tile, region, score, overlap, pars, focals, strategies, DEB)
        if uniform(0,1) < p:
            new_strategy.append(tile)
#    if DEB:
#        imprime_region(region)
#        print(f"Puntaje: {score}")
#    print("")
    return new_strategy
```
