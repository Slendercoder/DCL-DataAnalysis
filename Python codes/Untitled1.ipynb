```python
import pandas as pd
import seaborn as sns
```

Load data from sweeping **shaky hand parameter**

```python
df_list = []
for p in [0, 3, 6]:
    archivo = '../Data/output-shaky-' + str(p) + '-0.csv' 
    try:
        aux = pd.read_csv(archivo)
        aux['Parameter'] = p
        df_list.append(aux)
    except:
        pass

data = pd.concat(df_list)
data.head()
```
