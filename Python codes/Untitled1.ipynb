```python
import pandas as pd
import seaborn as sns
```

```python
df_list = []
for p in [0, 3, 6]:
    archivo = '../Data/output-shaky-' + str(p) + '-0.csv' 
    try:
        aux = pd.read_csv(archivo)
        aux['Shaky'] = p
        df_list.append(aux)
    except:
        pass

data = pd.concat(df_list)
data.head()
```

```python
sns.lineplot(x=data['Round'], y=data['DLIndex'], hue=data['Shaky'])
```

```python
sns.lineplot(x=data['Round'], y=data['Ac_Score'], hue=data['Shaky'])
```
